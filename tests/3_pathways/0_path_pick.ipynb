## Pathes picking

Here we check available pathes and select three for future analysis

````python
import csv
````

Let's first check all available pathes in KEGG, and their length:

````python
lines = None
with open("../data/c2.cp.kegg.v2023.1.Hs.symbols.gmt", "r") as in_gmt:
    lines = in_gmt.readlines()
lines = list(map(lambda x: x.split(), lines))

path2genes = {path[0]: path[2:] for path in lines}
path2size = {p: len(path2genes[p]) for p in path2genes}
path2size = sorted(path2size.items(), key=lambda x: x[1])
print(f"Length: {len(path2size)}")
path2size
````

### Selection and saving

````python
paths = [
    "KEGG_STEROID_BIOSYNTHESIS",
    "KEGG_PPAR_SIGNALING_PATHWAY",
    "KEGG_FOCAL_ADHESION"
]

names = ['small','medium', 'big']
````

````python
for path, name in zip(paths, names):
    with open(f'in_data/path_{name}.txt', 'w') as f:
        write = csv.writer(f)
        write.writerow([path,])
        print(f"{name}: {path}, {len(path2genes[path])}")
````
